```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris=load_iris()

```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target']= iris.target
df
```

```python
iris.target_names
```

```python
df[df.target==2].head()
```

```python
df['flower_name']=df.target.apply(lambda x: iris.target_names[x])
df
```

```python
%matplotlib inline
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
df2
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='pink',marker='*')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='green',marker='*')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='purple',marker='*')
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title('Sepallength v Sepalwidth')
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='red',marker='*')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='green',marker='.')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='purple',marker='+')
plt.xlabel("petal length")
plt.ylabel("petal width")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(["target","flower_name"],axis='columns')
x
```

```python
y=df.target
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
len(x_train)
```

```python
y_train
```

```python
from sklearn.svm import SVC
model= SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
