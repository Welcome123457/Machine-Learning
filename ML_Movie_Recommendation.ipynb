```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv(r"C:\Users\PAVAN G\Downloads\dataset.csv")
movies
```

```python
movies["tags"]=movies["genre"]+movies["overview"]
```

```python
movies.head(5)
```

```python
new_df=movies[["id","title","genre","overview","tags"]]
new_df
```

```python
new_df=new_df.drop(["genre","overview"],axis="columns")
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features=10000, stop_words="english")
cv
```

```python
vector=cv.fit_transform(new_df["tags"].values.astype("U")).toarray()
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
similarity
```
