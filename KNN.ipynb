```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sn
iris=load_iris()

```

```python
dir(iris)
```

```python
iris.feature_names
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target']= iris.target
df
```

```python
df['flower_name']=df.target.apply(lambda x: iris.target_names[x])
df
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='pink',marker='*')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='green',marker='*')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='purple',marker='*')
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title('Sepallength v Sepalwidth')
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='red',marker='*')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='green',marker='.')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='purple',marker='+')
plt.xlabel("petal length")
plt.ylabel("petal width")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(["target","flower_name"],axis='columns')
x
y=df.target
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
len(x_train)
```

```python
len(y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kc=KNeighborsClassifier(n_neighbors=3)
kc.fit(x_train,y_train)
```

```python
kc.score(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
y_p=kc.predict(x_test)

cm=confusion_matrix(y_test,y_p)
cm
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)

plt.xlabel("Predicted")
plt.ylabel("Truth")
```
